# Visualizing Loans Awarded by Kiva

In this project I'll visualize insights using a dataset from <a href = "https://www.kaggle.com/fkosmowski/kivadhsv1" target = "_blank">Kaggle</a>. The dataset contains information about loans awarded by the non-profit <a href = "https://www.kiva.org/" target = "_blank">Kiva</a>. 

Using Seaborn, I'll explore the average loan amount by country using aggregated bar charts. I'll also visualize the distribution of loan amounts by project type and gender using box plots and violin plots.

## Step 1: Importing Necessary Python Modules

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

## Step 2: Ingesting The Data
Loading **kiva_data.csv** into a DataFrame called `df`. Then, quickly inspecting the DataFrame using `.head()`.

```python
df = pd.read_csv('kiva_data.csv')
print(df.head())
```

## Step 3: Examine The Data

You can find **kiva_data.csv** in the project folder.

### Overview of the dataset:

Each entry (row) in the dataset represents a loan that Kiva awarded to a particular project. The `loan_amount` column shows the amount (in U.S. dollars) awarded to the project. The `activity` column has the category type that the project falls under. The `country` column is the country where the project is located. The `gender` column represents the gender of the primary person who applied for the loan. 


Printing the first 25 rows of `df` using `.head()`

```python
print(df.head(25))
```

## Step 4: Bar Charts

Creating a bar plot using Seaborn to visualize the average size of Kiva loans given to projects, by country.

I've set up the figure used to plot the bar plot on. The `f` variable gives us access to the figure and `ax` gives us access to the axes.

```python
# Creates the figure, I am only using this syntax to modify the y-axis ticks later
f, ax = plt.subplots(figsize=(15, 10))

sns.barplot(data=df, x="country", y="loan_amount")
plt.show()
```

### Adding `$` units

Run the code in the cell below to see the `$` in action.

```python
import matplotlib.ticker as mtick

# Creates the figure, I am only using this syntax to modify the y-axis ticks below
f, ax = plt.subplots(figsize=(15, 10))
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick) 

sns.barplot(data=df, x="country", y = "loan_amount")
plt.show()
```

## Step 5: Using `hue` In The Visualization

We can visualize even more data on one bar plot by visualizing the loan amount by country, and "nesting" by gender. Adding the `hue` parameter to my `sns.barplot()` and setting it so that the visualization includes the nested category of gender.

```python
f, ax = plt.subplots(figsize=(15, 10))

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

sns.barplot(data=df, x="country", y = "loan_amount", hue="gender")
plt.show()
```

## Step 6: Styling


Setting a different color palette using `sns.set_palette()`. 

Setting the plot background style using `sns.set_style()`.

Setting the title using `plt.title("")`.


```python
# Sets color palette
sns.set_palette("Accent")

# Sets style
sns.set_style("whitegrid")

# Creates figure and axes (no need to use the previous syntax, as the y-label ticks aren't going to be formatted)
plt.figure(figsize=(25, 15))

# Adds a title
plt.title("Average Loan Amounts Given by Kiva, by country, by gender (in $ USD)")

# Use Seaborn to create the bar plot
sns.barplot(data=df, x="country", y="loan_amount", hue="gender")
plt.show()
```

## Step 7: Box Plots With Kiva Data

So far we have visualized the average size of loans by country using bar charts; now we are going to make a box plot to compare the distribution of loans by country.

```python
plt.figure(figsize=(16, 10))
sns.boxplot(data=df, x="country", y="loan_amount")
plt.show()
```

## Step 8: Box Plot by Activity

Instead of visualizing the loan amount by *country*, we use `sns.boxplot()` to plot the loan amount by *activity*.

```python
plt.figure(figsize=(16, 10))

sns.boxplot(data=df, x="activity", y="loan_amount")
plt.show()
```

## Step 9: Violin Plots

```python
plt.figure(figsize=(16, 10))

sns.violinplot(data=df, x="activity", y = "loan_amount")
plt.show()
```

### Creating a violin plot that visualizes the distribution of loan amount by country.

```python
plt.figure(figsize=(16, 10))

sns.violinplot(data=df, x="country", y = "loan_amount")
plt.show()
```

## Step 10: Split Violin Plots

Using the `hue` and `split` parameters with `sns.violinplot()` to visualize the distribution of loan amount by country, split by gender. 

```python
sns.set_palette("Spectral")
plt.figure(figsize=(18, 12))

sns.violinplot(data=df, x="country", y = "loan_amount", hue="gender", split=True)
plt.show()
```
